```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
over20_df = vine_df[vine_df['total_votes'] >= 20]
over20_df.head(10)
```

```python
helpful_df = over20_df[over20_df['total_votes'] <= 2 * over20_df['helpful_votes']]
helpful_df.head(10)
```

```python
paid_df = helpful_df[helpful_df['vine'] == 'Y']
paid_df.head()
```

```python
unpaid_df = helpful_df[helpful_df['vine'] == 'N']
unpaid_df.head()
```

```python
paid_total = paid_df['review_id'].count()
unpaid_total = unpaid_df['review_id'].count()
```

```python
paid_5s_df = paid_df[paid_df['star_rating'] == 5]
unpaid_5s_df = unpaid_df[unpaid_df['star_rating'] == 5]
```

```python
paid_5s = paid_5s_df['review_id'].count()
unpaid_5s = unpaid_5s_df['review_id'].count()
unpaid_5s
```

```python
paid_percent = (100 * paid_5s / paid_total)
unpaid_percent = (100 * unpaid_5s / unpaid_total)
```

```python
summary = (
    f"-------------------------\n"
    f"Total Paid Reviews: {paid_total:,}\n"
    f"5-Star Paid Reviews: {paid_5s:,}\n"
    f"5-Star Percentage: {paid_percent:.1f}%\n"
    f"-------------------------\n"
    f"Total Unpaid Reviews: {unpaid_total:,}\n"
    f"5-Star Unpaid Reviews: {unpaid_5s:,}\n"
    f"5-Star Percentage: {unpaid_percent:.1f}%\n")
print(summary)
```
